```python
pitstop_df["driver_name"]=driver_df["forename"]
pitstop_df.head()
```

```python
def id_to_name(id):
    row = driver_df[driver_df['driverId'] == id]
    if not row.empty:
        return row.iloc[0].forename+" " +row.iloc[0].surname
    else:
        return "Unknown"
```

```python
pitstop_df["driver_name"]=pitstop_df["driverId"].apply(id_to_name)
```
