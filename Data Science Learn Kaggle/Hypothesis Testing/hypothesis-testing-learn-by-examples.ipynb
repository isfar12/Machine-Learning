# Basic hypothesis test walkthrough

### 1. Test description

Question:

Does our dice roll a 3 or less 50% of the time and a 4+ 50% of the time?

Method:

We roll a dice 10 times noting the results. We write 1 if the roll is greater or equal to 4 and 0 otherwise.

            Number rolled       >=4
               
    roll 1       1              0

    roll 2       1              0
    
    roll 3       4              1

    roll 4       5              1

    roll 5       1              0

    roll 6       3              0
    
    roll 7       3              0

    roll 8       3              0

    roll 9       1              0

    roll 10      6              1

### 2. Set a null hypothesis

First we assume that it does roll a 4+ 50% of the time, this is our **null hypothesis** usually notated as $H_0$. Therefore our **alternative** hypothesis must be that the dice does not roll a 4+ 50% of the time.

### 3. Set a critical value

We set a critical value of $\alpha = 0.05$. This is equivalent to saying, we will only reject the null hypothesis if our results are so extreme that the probability of obtaining them (or more extreme results) is less than 5% whilst assuming the null hypothesis to be true. Note that the critical value is totally arbitrary depending on what goal you re trying to achieve.

### 4. Find the p-value

The p-value is the probability of seeing results at least as extreme as the results we've obtained whilst assuming the null hypothesis to be true. We rolled a 4+ 3 times, so we need to find the probability of rolling a 4+ 3 or less times. This means finding the probability of rolling a 4+  0,1,2 and 3 times and adding them up.

0) The probability of roll a 4+ 0 times is $0.5^{10}$.

1) The probability of roll a 4+ 1 time is ${10 \choose 1} \cdot 0.5^{10}$. 

2) The probability of roll a 4+ 2 times is ${10 \choose 2} \cdot 0.5^{10}$.

3) The probability of roll a 4+ 3 times is ${10 \choose 3} \cdot 0.5^{10}$.

Adding all of these up we get $\sum {10 \choose i} 0.5^{10} = 0.5^{10} \sum {10 \choose i} = \dfrac{11}{64} = 0.171875$

<i> We found the p-value explicitly here using the binomial distribution. Our test statistic was 3 i.e the number to input into our probability distribution function. In most cases the type of test is dependant on the type of distribution that's needed to calculate p from the test statistic.</i> 

### 5. Conclusion

If this were a fair dice, it would get a result at least as extreme as ours 17% of the time which is above the critical value and so we accept the null hypothesis.

# Chi squared

Chi-squared example 1: Goodness of fit
<a id='Goodness of fit'></a>

### 1. Test description

We roll a dice 60 times noting the results to test whether ot not it's fair.

    dice face     1     2     3     4     5     6

    count         9     8     10    14    8     11

### 2. Set null hypothesis

$H_0$: Our dice is fair.

$\implies p(i) = 1/6$  for  $i \in \{1, 2, 3, 4, 5, 6\}$ 

### 3. Set critical value

$\alpha = 0.05$ is our critical value.

### 4. Find the p-value

We need to calculate the $\chi^2$ statistic using the formula $\chi^2 = \sum \dfrac{(O_i - E_i)^2}{E_i}$.

Here $O_i$ is the observed number of rolls of $i$ and $E_i$ is the expected number.

If the dice were fair we'd expect 60/6 = 10 appearances  for each face of the dice.

Now we can calculate our test statistic $\chi^2$.

$\chi^2 = \frac{(10 - 10)^2}{10} + \frac{(14 - 10)^2}{10} + \frac{(6 - 10)^2}{10} + \frac{(9 - 10)^2}{10} + \frac{(13 - 10)^2}{10} + \frac{(8 - 10)^2}{10} = \frac{0}{10} + \frac{16}{10} + \frac{16}{10} + \frac{1}{10} + \frac{9}{10} + \frac{4}{10} = \frac{42}{10} = 4.2$

We look to our chi-squared table to find p.

The rows of the table represent degrees of freedom which can be calculated from the example as n-1 where n is the number of different possible roll values.

So in this case the degrees of freedom is 5.

<img style="float:left" src="https://i.imgur.com/f3zHaj4.png" alt="Chi square table" style="width:435px;height:366px;">
<div style="clear:left">

A $\chi^2$ value of at least 11.07 is needed for a p-value of 0.05 or less, so we can safely say that our p-value is well over 0.05.
    
### 5. Conclusion

Given that $p > \alpha = 0.05$, there is not enough evidence to reject the null hypothesis.

Chi-squared example 2: Homogeneity
<a id='Homogeneity'></a>

### 1. Test description

We have two 3 sided dice and want to test whether they each roll a 3 as often as each other.

Rolling dice 1 and 2 60 times each.

                1     2     3

    dice 1     17    25    18 

    dice 2     15    22    23
    
### 2. Set null hypothesis


$H_0$: Dice 1 rolls a 3 as often as dice 2 rolls a 3.

$\implies p(x_1 = 3) = p(x_2 = 3)$

### 3. Set critical value

$\alpha = 0.05$ is our critical value.

### 4. Find the p-value

We need to calculate the $\chi^2$ statistic using the formula $\chi^2 = \sum\dfrac{(O_i - E_i)^2}{E_i}$

Here $O_i$ is the observed number of rolls of $i$ and $E_i$ is the expected number.

Assuming the null hypothesis is true, the expected frequency of 3 on dice 1 is 23 expected occurences since that's how many we rollled for dice 2.

$\chi^2 = \frac{(23 - 18)^2}{23} = 1.08695652$

Looking to our table with degrees of freedom of (columns -1)\*(rows - 1) = (3-1)\*(2-1) = 2

<img style="float:left" src="https://i.imgur.com/dL3CXZs.png" alt="Chi square table" style="width:435px;height:366px;">
<div style="clear:left">

A $\chi^2$ value of at least 5.991 is needed for a p-value of 0.05 or less, so we can safely say that our p-value is well over 0.05.

### 5. Conclusion

Given that $p > \alpha = 0.05$, there is not enough evidence to reject the null hypothesis.

Chi-squared example 3: Independence

<a id='Independence'></a>

### 1. Test description

When rolling a 6 sided dice, we set the random variable X as: rolling a prime number and the random variable Y as: rolling an odd number.

We'd like to test whether the two random variables are indpendent.

We roll the dice 20 times with the following results.

Three 1s,  two 2s, four 3s, four 4s, two 5s, five 6s.
 
               even   odd   total

    prime        2     6      8 

    not-prime    9     3      12

    total        11    9      20

### 2. Set null hypothesis

$H_0:$ X and Y are independant.

### 3. Set critical value

$\alpha = 0.05$ is our critical value.

### 4. Find the p-value

If these two random variables are independant, then p(X=x, Y=y) = p(X=x) \cdot p(Y=y).

From the above property of independence, we can find the expected counts.

p(X=prime) = 8/20, p(X=not-prime) = 12/20

p(Y=odd) = 9/20, p(Y=even) = 11/20

So the expected value of rolling a prime and even number is (8/20)\*(11/20)\*20 = 88/20 = 22/5.

Computing $\chi^2$.

$\chi^2 = \dfrac{(2-22/5)^2}{22/5} + \dfrac{(6-18/5)^2}{18/5} + \dfrac{(11-33/5)^2}{11/5} + \dfrac{(3-27/5)^2}{27/5} = \dfrac{76}{11} = 6.909090$

Looking to our table with degrees of freedom of (columns -1)\*(rows - 1) = (2-1)\*(2-1) = 1

<img style="float:left" src="https://i.imgur.com/xQEJZTV.png" alt="Chi square table" style="width:435px;height:366px;">
<div style="clear:left">
    
$\chi^2$ is greater than 3.841 and so $p < 0.05$. In fact from the table we see that $p < 0.01$.

### 5. Conclusion

Since $p < \alpha = 0.05$ we must reject the null and conclude that X and Y are no independent.

# Z-test

Z-test example 1:

### 1. Test description

The distribution of a fair dice has a mean of 3.5 and a standard deviation of $\dfrac{\sqrt{105}}{6}$.

We have a sample X of n = 100 rolls of a dice and we would like to know if it's considered fair.

The sample has a mean of 3.11 with standard deviation 1.3808.

### 2. Set null hypothesis

$H_0$: The dice is fair.

$\implies$: The mean is $3.5$ with standard deviation $\dfrac{\sqrt{105}}{6}$.


### 3. Set critical value

We will set our critical value at $\alpha = 0.1$ for a two-tailed test.

### 4. Find the p-value

The standard error of the mean is: $SE = \dfrac{\sigma}{\sqrt{n}} = \dfrac{\sqrt{105}}{6} \cdot \dfrac{1}{\sqrt{100}} = \dfrac{\sqrt{105}}{60}$.

We now calculate the z-score i.e the distance of the sample mean from the population mean in units of standard error.

$z = \dfrac{E(X) - \mu}{SE} = (3.11 - 3.5) \cdot \dfrac{60}{\sqrt{105}} = -2.2836$

Since the normal distribution is symetrical, the probability of obtaining 2.2836 and -2.2836 is the same.

<img style="float:left" src="https://i.imgur.com/BT2s6l6.png" alt="Z-table" style="width:435px;height:366px;">
<div style="clear:left">
   
$Z$ is less than -1.645 (greater than 1.645 by symetry) and so $p < 0.05$ for a one-tailed test therefore $p < 0.1$ for a two-tailed test.

Looking up the p-value in the z table yields $p < 0.05$ which for a two-tailed test must be doubled to p = 0.0226.

### 5. Conclusion

The p-value is extremely low in this test so we must reject the null hypothesis and assume the dice is unfair.

# T-test

<a id='Two-sampled'></a>

### 1. Test description

T-test example 1: Two-sample T-test

A friend has lent us two dice that are supposedly weighted.

To check which dice rolls higher, we roll both dice 50 times each and note the results.

               mean   standard deviation   n

    dice 1     3.9           1.2           50 

    dice 2     4.3            1            50
    
Dice 2 seems to rolls higher on average, we would like to test whether this is true.

### 2. Set null hypothesis

$H_0$: Both dice have the same mean roll value.

$\implies \mu_1 = \mu_2 \implies \mu_1 - \mu_2 = 0$

### 3. Set critical value

Our critical value is $\alpha = 0.05$.

### 4. Find the p-value

T-statistic = $\dfrac{\mu_1 - \mu_2}{\sqrt{\dfrac{s_1^2}{n_1}+\dfrac{s_2^2}{n_2}}} = \frac{4.3 - 3.9}{\sqrt{\dfrac{1.2^2}{50}+\dfrac{1^2}{50}}} = 1.8107$

Degrees of freedom is (50-1)+(50-1) = 98

Finding p from the T-table:

<img style="float:left" src="https://i.imgur.com/T3WJgaJ.png" alt="T-table" style="width:683px;height:171px;">
<div style="clear:left">
    
We find $p$ to be between 0.025 and 0.05.

### 5. Conclusion

Given that $p < \alpha = 0.05$ we must to reject the null hypothesis and assume that dice 2 does roll higher than dice 1.

T-test example 2: Paired T-test

<a id='Paired'></a>

### 1. Test description

Does the temperature of a dice change it's mean value?

We roll 10 dice in a cold room and note the values.

We then increase the temperature of the room and reroll the dice and note the results.

               cold        hot      difference
               
    dice 1       1          3          -2

    dice 2       1          6          -5
    
    dice 3       4          5          -1

    dice 4       5          2           3

    dice 5       1          6          -5

    dice 6       3          3           0
    
    dice 7       3          1           2

    dice 8       3          4          -1

    dice 9       1          4          -3

    dice 10      6          4           2


The difference column has a mean of -1 with standard devation 2.68.

### 2. Set null hypothesis

$H_0$: There is no difference in mean values when the temperature is changed.

$\implies \mu_{cold} = \mu_{hot} \implies \mu_{difference} = 0$

### 3. Set critical value

Let's choose a standard critical value of $\alpha = 0.05$.

### 4. Find the p-value

Now we can calulate our standard error for the difference column.

SE = $\dfrac{s}{\sqrt{n}} = \dfrac{2.68}{\sqrt{10}} = 0.847$

And finally the T-statistic for finding our p-value.

T-statistic = $\dfrac{\mu - 0}{SE} = \dfrac{-1}{0.847} = -1.181$

Degrees of freedom is 10-1 = 9

Finding p from the T-table:

<img style="float:left" src="https://i.imgur.com/jB3SGXs.png" alt="T-table" style="width:683px;height:171px;">
<div style="clear:left">
    
By symmetry P(T<-1.181) = P(T>1.181).
    
We find $p > 0.05$

### 5. Conclusion

$p > \alpha = 0.05$ and so there is not enough evidence to reject the null hypothesis.

# ANOVA

ANOVA test example 1:

### 1. Test description

Similarly to our paired T-test experiment, we'd like to test the perfomence of dice in different temperatures.

Only this time there will be 3 different temperatures.

Here are the resuts of m = 3 different categories, with n = 30 samples.

               cold        medium      hot
               
    dice 1       4          4           5

    dice 2       2          3           5
    
    dice 3       1          6           1

    dice 4       6          5           4 

    dice 5       4          6           5

    dice 6       4          1           4
    
    dice 7       2          5           1

    dice 8       3          5           5

    dice 9       2          5           2

    dice 10      5          1           4
    
    mean         3.3        4.1         3.6
    
    SD           1.487      1.758       1.562
    
The total mean is $\bar{X} = 3.67$.

### 2. Set null hypothesis

Our null hypothesis is still that there is no change in the rolls with varying temperature.

### 3. Set critical value

$\alpha = 0.05$ is our critical value.

### 4. Find the p-value

Calculating the F-statistic, we begin with the variation **between** the groups.

$SS_{between} = \sum n_i (\bar{X_i} - \bar{X})^2 = 10 \cdot (3.3 - 3.67)^2 + 10 \cdot (4.1 - 3.67)^2 + 10 \cdot (3.6 - 3.67)^2 = 3.267$

$df_{between} = m - 1 = 3 -1 = 2$

$MS_{between} = \dfrac{SS_{between}}{df_{between}} = \dfrac{3.267}{2} = 1.634$

And now the variation **within** the groups (a bit like total variation).

$SS_{within} = \sum (X_i - \bar{X_j})^2 = (4-3.3)^2 + (2 - 3.3)^2 + ... + (2 - 3.6)^2 + (4-3.6)^2 = 77.4$

$df_{within} = n - m = 30 - 3 = 27$

$MS_{within} = \dfrac{SS_{within}}{df_{within}} = \dfrac{77.4}{27} = 2.86$

The F-statistic is the variation between by the variation within (total).

$F = \dfrac{MS_{between}}{MS_{within}} = \dfrac{1.634}{2.86} = 0.571$

We can find p by checking the F-table with $\nu_1 = 3-1 =2$ and $\nu_2 = 10-1 = 9$.

<img style="float:left" src="https://i.imgur.com/fnZGnEm.png" alt="F-table" style="width:765px;height:183px;">
<div style="clear:left">

$F$ is far lower than 4.26 and so p is much higher than 0.05.
    
### 5. Conclusion

Since $p > \alpha = 0.05$ there is no reason to reject the null hypothesis.
