```python
plt.figure(figsize=(9,6))
plt.style.use("seaborn-v0_8-darkgrid")
plt.title('Import from Worldwide to Uk over Years',color="black")
plt.plot(df2["Year"],df2["Europe"],"b",df2["Year"],df2["Asia"],"r",df2["Year"],df2["North America incl. W Indies to 1972"],"g",df2["Year"],df2["South and Central America"],"y",df2["Year"],df2["Australia"],"k")
plt.legend(["Europe","Asia","North America incl. W Indies to 1972","South and Central America","Australia"],loc="upper right")
plt.grid(True)
plt.show()
```
