```python
newdf=df.copy()
newdf.head()
```

```python
#data converstion and fixing types
def wage_fix(data):
    return int(str(data).replace("€","").replace("K",""))*1000
newdf["Wage"]=newdf["Wage"].apply(wage_fix)
newdf["Wage"].head()
```

```python
def clause_fix(data):
    return float(str(data).replace("€","").replace("M","").replace("K",""))*1000000
newdf["Release Clause"]=newdf["Release Clause"].apply(clause_fix)
newdf["Release Clause"].head()
```
