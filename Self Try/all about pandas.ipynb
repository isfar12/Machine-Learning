# aggregate and Grouping


```python
import pandas as pd
```

```python
pokemon=pd.read_csv("./Pandas Dataset/pokemon_data.csv")
pokemon.head()
```

```python
pok_type=pokemon.groupby(by="Type 1") # grouping dataset according to columns
pok_type
```

```python
print("Mean Value of Hp row ")
print(pokemon["HP"].mean()) ##getting mean of the entrie column of the dataset
print("Mean Value of Hp According to Group ")
print(pok_type["HP"].mean()) ##getting mean of groups as created in pok_type group
```

```python
pok_type.agg({'Attack':['mean','sum','count','max'],'Sp. Atk':['mean','sum','count','max']}).head() # agg is used to see multiple queries ata  time
```

```python
pok_type.describe().head()
```

```python
pok_type.agg({"Speed":"mean"}).head() 
```
