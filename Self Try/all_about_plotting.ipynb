# Modifying plot function

```python
y=[1,2,3,4,5,6,8,9]
plt.plot(y,"r+-") #string format is " color marker line"
```

```python
plt.plot(y,"r+-",antialiased=True,linewidth=10,linestyle="--",color="blue",marker=".",markerfacecolor="green",markeredgecolor="red") 
# antialaised used to show smooth line
# linewidth used to define the width of line
#linestyle means hw the line will be shown
#color= changes the color of line and marker
# marker = the type of the marker
#makerfacecolor=change the inside color of marker
#markeredgecolor=change the outline color of marker
```

```python
plt.plot(y,alpha=.5) #alpha means tranparent view, range 0-1
```

```python
plt.plot(y,zorder=1,linewidth=10) 
plt.plot([pow(i,3) for i in y],zorder=2,linewidth=10)
plt.show()
#zorder says if there are multiple line overlapping which one should be on top and which one should be next, by indexing 1,2,3 etc
```

# xlabel() ylabel() function

```python
plt.plot(y)
plt.xlabel("X values",{"color":"red","font":"Times New Roman","style":"italic","backgroundcolor":"yellow"},labelpad=20,size=20,weight="bold")

'''there are 2 ways to write attributes. one to use dictionary, another to use keyword like after'''
#color: color of the label
# font: font type
# style: font style
# backgroundcolor: label background
#labelpad: the height of the label, -ve value means up, +ve means lower
#size: size of font
#weight: the weight, light bolod etc
```

# Title() grid() function

```python
plt.plot(y)
plt.title("Simple Plot",loc="right",size=20,pad=50) #pad means space up from the bar, and loc means location,right,left,centre
plt.xlim(0,9) #range from which value to which show in x axis
plt.ylim(0,12)
plt.grid(True,color="red")
```

# legend() function

```python
y=[2,4,6,8,10,12,14,16]
plt.plot(y,"r",[i**2 for i in y])
plt.legend(["y","y square"],loc="center right",title="Line indicator")#we use list to specify the order, which plot will take which string, and loc for
# to know the position of the legend
plt.grid()
plt.show()
```

# Normal Distribution

```python
import numpy as np
x=np.random.normal(20,1.5,1500)
plt.hist(x,density=True,bins=40,edgecolor="black")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0
std_dev = 1
num_samples = 1000

# Generate random data following a normal distribution
data = np.random.normal(mean, std_dev, num_samples)

# Create a histogram of the data
plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Plot the PDF of the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')

# Show the plot
plt.show()

```
