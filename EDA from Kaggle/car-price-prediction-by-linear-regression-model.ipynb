<a id="step3-7"></a>
# <b><span style='color:#00008B'>Step 3.7 |</span><span style='color:#00008B'> Pie and Bar charts for 'categorical' features </span></b>
⬆️ [Tabel of Contents](#contents_tabel)

```python
Categorical_Features = ['Fuel_Type',
                        'Seller_Type',
                        'Transmission',
                        'Owner']

Categorical_Features_Title = ['<b> Fuel_Type </b>',
                              '<b> Seller_Type </b>',
                              '<b> Transmission </b>',
                              '<b> Owner </b>']

Hole_Size = [0.5, 0.5, 0.6, 0.4]

i = 0

for Feature in Categorical_Features:
    
    df_study = df_categorical[Feature].value_counts().reset_index()
        
    specs = [[{'type':'pie'}, {"type": "bar"}]]
    fig = make_subplots(rows=1, cols=2, specs=specs, shared_yaxes = True, subplot_titles=['Pie Chart', 'Bar Chart'])
    
    fig.add_trace(go.Pie(
                        labels = df_study[Feature], 
                        values = df_study['count'],
                        name='',
                        hole = Hole_Size[i],
                        marker_colors = ['#00BFFF', '#FF1493','#00FF00'],
                        textinfo='label+percent',
                        ), 
                        1, 1)  
    
    fig.update_traces(hoverinfo="label+percent+name", textfont_size=20)
    
    trace1 = go.Bar(
                    x=df_study[Feature],
                    y=df_study['count'],
                    text=df_study['count'],
                    name='',
                    textposition='auto',
                    textfont=dict(size=16),
                    marker=dict(color=['#00BFFF', '#FF1493','#00FF00'],)
                    )
    
    
    fig.append_trace(trace1, 1,2)

    if Feature == 'Fuel_Type':
        
        fig.update_layout(title={
                                'text': '<b> Pie and Bar Charts for the Count of any Type in each Categorical Feature </b>',
                                'font': {'size': 18, 'color': 'black'}
                                 }
                         )

    
    fig.update_layout(
                        showlegend=False, 
                        height=500,
                        annotations=[dict(text= Categorical_Features_Title[i], x=0.225, y=0.45, font_size=20, showarrow=False)]
                      ) 
                     
    fig.update_xaxes(tickfont_family="Arial Black")
    fig.update_yaxes(tickfont_family="Arial Black")
    
    fig.show()

    i += 1
```

<a id="step4-2-1"></a>
# <b><span style='color:#00008B'>Step 4.2.1 |</span><span style='color:#00008B'> Adding effective power 2 of features </span></b>
⬆️ [Tabel of Contents](#contents_tabel)

```python
Order=2
TestSize=0.2
RandomState=1

R2 = R2_FM_1

Column_Name_List=[]

X_FM_2 = df_FirstModel.loc[:, df_FirstModel.columns != 'Selling_Price']
y_FM_2 = df_FirstModel['Selling_Price'].values.reshape(-1,1)

for column in X_FM_1:
    
    Column_Name = column + str(Order)
    Column_Value = X_FM_2[column] ** Order
    Column_index = X_FM_2.columns.get_loc(column) + 1
    X_FM_2.insert(Column_index, Column_Name, Column_Value, True)
   
    X_train_FM_2, X_test_FM_2, y_train_FM_2, y_test_FM_2 = train_test_split (X_FM_2, y_FM_2, test_size=TestSize, random_state=RandomState)
    First_Model_2 = LinearRegression()
    First_Model_2.fit(X_train_FM_2 , y_train_FM_2)
    y_pred_FM_2 = First_Model_2.predict(X_test_FM_2)
    y_pred_FM_2[y_pred_FM_2 < 0] = 0
    R2_New = metrics.r2_score(y_test_FM_2, y_pred_FM_2)
    
    print('\n', start + 'R2 for adding',Column_Name,'is: ', R2_New)

    if R2_New > R2:
        R2 = R2_New
        Column_Name_List.append(Column_Name)
    else:
        X_FM_2 = X_FM_2.drop(columns = Column_Name)
         
print ('\n', start + 'Final R2 is: ', R2,'\n')

R2_FM_2 = R2

new_dimensions_count = 0
for col in Column_Name_List:
    print('\n', start + 'New added dimension is: ', col)
    new_dimensions_count += 1

print('\n', start + 'The number of added new and effective dimensions is: ', new_dimensions_count)
```

```python
TestSize=0.2
RandomState=1

R2 = R2_FM_2

Column_Name_List=[]

X_FM_3 = X_FM_2.copy()
y_FM_3 = y_FM_2.copy()

for column_1 in X_FM_2:
    
    for column_2 in X_FM_2:

        if column_1 != column_2:
        
            Column_Name = column_1 + '.' + column_2

            Column_Name_Reverse = column_2 + '.' + column_1

            if Column_Name_Reverse not in Column_Name_List:
            
                Column_Value = X_FM_3[column_1] * X_FM_3[column_2]
                X_FM_3.insert(loc=len(X_FM_3.columns), column=Column_Name, value=Column_Value)

                X_train_FM_3, X_test_FM_3, y_train_FM_3, y_test_FM_3 = train_test_split (X_FM_3, y_FM_3, test_size=TestSize, random_state=RandomState)
                First_Model_3 = LinearRegression()
                First_Model_3.fit(X_train_FM_3 , y_train_FM_3)
                y_pred_FM_3 = First_Model_3.predict(X_test_FM_3)
                y_pred_FM_3[y_pred_FM_3 < 0] = 0
                R2_New = metrics.r2_score(y_test_FM_3, y_pred_FM_3)

                print('\n', start + 'R2 for adding',Column_Name,'is: ',R2_New)
            
                if R2_New > R2:
                    R2 = R2_New
                    Column_Name_List.append(Column_Name)
                else:
                    X_FM_3 = X_FM_3.drop(columns = Column_Name)
         
print ('\n', start + 'Final R2 is: ', R2,'\n')

new_dimensions_count = 0
for col in Column_Name_List:
    print('\n', start + 'New added dimension is: ',col)
    new_dimensions_count += 1

print('\n', start + 'The number of added new and effective dimensions is: ', new_dimensions_count)
```

# <b><span style='color:#00008B'>Step 5.1.1 |</span><span style='color:#00008B'> Scaling the input data </span></b>
⬆️ [Tabel of Contents](#contents_tabel) 

```python
# Present_Price = 11.23
# Kms_Driven = 42000
# Owner = 1
# Age = 10 
# Fuel_Type = 11 # Petrol 
# Seller_Type = 21  # Dealer
# Transmission = 31 # Manual

Inputs_Real = [11.23, 42000, 11, 21, 31, 1, 10]
Inputs_Name = ['Present_Price', 'Kms_Driven', 'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner', 'Age']
Inputs_Scaled = []
i = 0

for feature in Inputs_Name:
    
    minimum_value = df4[feature].min()
    range_value = df4[feature].max()-df4[feature].min()
    Inputs_Scaled.append((Inputs_Real[i] - minimum_value)/range_value)
    i += 1

scaled_data = {
               'Inputs Feature Name': Inputs_Name,
               'Scaled Value': Inputs_Scaled
              }

df1_scaled_inputs = pd.DataFrame(scaled_data)
 
display(df1_scaled_inputs)
```

<a id="step5-1-2"></a>
# <b><span style='color:#00008B'>Step 5.1.2 |</span><span style='color:#00008B'> Creating a model input dataframe from scaled input values</span></b>
⬆️ [Tabel of Contents](#contents_tabel) 

```python
Present_Price = df1_scaled_inputs.iat[0, 1]
Kms_Driven = df1_scaled_inputs.iat[1, 1]
Fuel_Type = 0 # Petrol 
Seller_Type = 0  # Dealer
Transmission = 0 # Manual
Owner = df1_scaled_inputs.iat[5, 1]
Age = df1_scaled_inputs.iat[6, 1] 

Model_Input = pd.DataFrame({
                        'Present_Price' : [Present_Price],
                        'Kms_Driven' : [Kms_Driven],
                        'Fuel_Type' : [Fuel_Type],
                        'Fuel_Type2' : [Fuel_Type**2],
                        'Fuel_Type3' : [Fuel_Type**3],
                        'Seller_Type' : [Seller_Type],
                        'Seller_Type2' : [Seller_Type**2],
                        'Transmission' : [Transmission],
                        'Owner' : [Owner],
                        'Age' : [Age],
                        'Age2' : [Age**2],
                        'Present_Price.Kms_Driven' : [Present_Price*Kms_Driven],
                        'Present_Price.Seller_Type' : [Present_Price*Seller_Type],
                        'Present_Price.Seller_Type2' : [Present_Price*(Seller_Type**2)],
                        'Present_Price.Transmission' : [Present_Price*Transmission],
                        'Present_Price.Owner' : [Present_Price*Owner],
                        'Kms_Driven.Owner' : [Kms_Driven*Owner],
                        'Fuel_Type.Seller_Type' : [Fuel_Type*Seller_Type],
                        'Fuel_Type.Seller_Type2' : [Fuel_Type*(Seller_Type**2)],
                        'Fuel_Type2.Seller_Type' : [(Fuel_Type**2)*Seller_Type],
                        'Seller_Type.Age' : [Seller_Type*Age],
                        'Seller_Type.Age2' : [Seller_Type*(Age**2)], 
                        'Seller_Type2.Owner' : [(Seller_Type**2)*Owner],
                        'Seller_Type2.Age' : [(Seller_Type**2)*Age],
                        'Transmission.Age' : [Transmission*Age],
                        'Transmission.Age2' : [Transmission*(Age**2)],
                        'Owner.Seller_Type' : [Owner*Seller_Type],
                        'Owner.Age' : [Owner*Age]
                        })

Model_Input
```
