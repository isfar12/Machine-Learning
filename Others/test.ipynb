```python
import numpy as np

# Define 1D arrays for x and y
x = np.linspace(-5, 5, 5)  # 5 points between -5 and 5
y = np.linspace(-5, 5, 5)

# Create meshgrid
X, Y = np.meshgrid(x, y)

# Print results
print("X:\n", X)
print("Y:\n", Y)
```

```python
import matplotlib.pyplot as plt

# Evaluate a function z = f(x, y) over the grid
Z = X**2 + Y**2

# Plot a contour plot
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title("Contour Plot of Z = X^2 + Y^2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
