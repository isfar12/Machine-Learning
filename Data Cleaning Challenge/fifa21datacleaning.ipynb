```python
import pandas as pd
import numpy as np
```

Importing Dataset

```python
fifa=pd.read_csv('/kaggle/input/fifa-21-messy-raw-dataset-for-cleaning-exploring/fifa21 raw data v2.csv')
fifa
```

Droping Unwanted Clomuns and Renameing Columns

```python
fifa.shape
```

```python
fifa.drop(columns=['Name','photoUrl','playerUrl','Nationality'],inplace=True)
```

```python
fifa.rename(columns={'LongName':'Name'},inplace=True)
fifa.head()
```

Remove the unnecessary newline characters from all columns that have them.

```python
fifa['Club']=fifa['Club'].str.replace('\n\n\n\n','')
fifa.head()
```

Convert the height and weight columns to numerical forms

```python
fifa['Height']=fifa['Height'].str.replace('cm','')
fifa.head()
```

```python
fifa['Height'].unique()
```

```python
def ToInteger(x):
    featInch=str(x).replace("\'",x) 
    return int(round(float(featInch[0])* 30.48 + float(featInch[1].replace('"',''))* 2.54))

fifa['Height']=fifa['Height'].apply(ToInteger)
```

```python
fifa['Height'].head()
```

Converting Weight

```python
fifa['Weight'].unique()

fifa['Weight']=fifa['Weight'].str.replace('kg','')
fifa['Weight']=fifa['Weight'].str.replace('lbs','')
```

```python
fifa.head()
```

'Value', 'Wage' and "Release Clause' are string columns. Convert them to numbers. For eg, "M" in value column is Million, so multiply the row values by 1,000,000, etc.

```python
def ConvertMoney(money:str):
    if '€' in money:
        money=money.replace("€","")
    if 'M' in money:
        return int(float(money.replace('M',''))*1000000)
    elif 'K' in money:
        return int(float(money.replace('K',''))*1000)

fifa['Value'] = fifa['Value'].apply(ConvertMoney)
fifa['Wage'] = fifa['Wage'].apply(ConvertMoney)
fifa['Release Clause'] = fifa['Release Clause'].apply(ConvertMoney)
fifa[['Value','Wage','Release Clause']]
```

Covert 'Hits' into numberical forms

```python
fifa['Hits'].unique()
```

```python
fifa['Hits'].apply(lambda x: float(x) if 'K' not in str(x)  else str(x).replace('K','')*1000)
```

Some columns have 'star' characters. Strip those columns of these stars and make the columns numerical

```python
fifa['IR']=fifa['IR'].apply(lambda x : str(x).split()[0])
fifa.head()
```

```python
fifa[fifa['Value']==fifa['Value'].max()]
```

Creating Separate Column Called "YearsOfPlaying"
